```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Titanic-Dataset.csv")
print(df.info())
print(df.head())

```

```python

fare = df[['Fare']]
print(fare.head())


class_age = df[['Pclass', 'Age']]
print(class_age.head())

survived_gender = df[['Survived', 'Sex']]
print(survived_gender.head())
```

```python
# 1. Fare > 100
fare_gt_100 = df[df['Fare'] > 100]
print(fare_gt_100.head())

# 2. Pclass = 1
first_class = df[df['Pclass'] == 1]
print(first_class.head())

# 3. Female under 18
female_under_18 = df[(df['Age'] < 18) & (df['Sex'] == 'female')]
print(female_under_18.head())
```

```python
# Embarked C or S
embarked_c_or_s = df[df['Embarked'].isin(['C', 'S'])]
print(embarked_c_or_s.head())

# Pclass in [1,2]
first_second_class = df[df['Pclass'].isin([1,2])]
print(first_second_class.head())
```

```python
# 1. Add fare per year column
df['fare_per_year'] = df['Fare'] / df['Age']

# 2. Subset where fare_per_year > 5
high_fare_age = df[df['fare_per_year'] > 5]

# 3. Sort descending
high_fare_age_srt = high_fare_age.sort_values(by="fare_per_year", ascending=False)

# 4. Select Name + fare_per_year
result = high_fare_age_srt[['Name', 'fare_per_year']]

# 5. Display
print(result.head(10))

```

```python
# 1. Add fare per class
df['fare_per_class'] = df['Fare'] / df['Pclass']

# 2. Subset adult males
adult_males = df[(df['Sex'] == 'male') & (df['Age'] >= 18)]

# 3. Sort descending
adult_males_srt = adult_males.sort_values(by="fare_per_class", ascending=False)

# 4. Select Name, Age, fare_per_class
result2 = adult_males_srt[['Name', 'Age', 'fare_per_class']]

# 5. Display
print(result2.head(10))
```

```python
# 1. Total fare overall
total_fare = df['Fare'].sum()

# 2â€“3. Total fare per class
fare_class_1 = df[df['Pclass'] == 1]['Fare'].sum()
fare_class_2 = df[df['Pclass'] == 2]['Fare'].sum()
fare_class_3 = df[df['Pclass'] == 3]['Fare'].sum()

# 4. Combine
fare_list = [fare_class_1, fare_class_2, fare_class_3]

# 5. Percent revenue per class
percentages = [(x / total_fare) * 100 for x in fare_list]

print("Percent fare revenue by class:")
print(f"First Class:  {percentages[0]:.2f}%")
print(f"Second Class: {percentages[1]:.2f}%")
print(f"Third Class:  {percentages[2]:.2f}%")
```

```python
# 1. Create age group column
def age_group(age):
    if age < 18:
        return "child"
    elif age < 65:
        return "adult"
    else:
        return "senior"

df['age_group'] = df['Age'].apply(age_group)

# 2. Total passengers
total_passengers = len(df)

# 3. Count per group
group_counts = df['age_group'].value_counts()

# 4. Percent of each group
percent_group = (group_counts / total_passengers) * 100

# 5. Display
print(percent_group)
```
