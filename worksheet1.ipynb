```python
import numpy as np

print("Problem 1: Array Creation")

empty_array = np.empty((2, 2))
print("1. Empty array (2x2):\n", empty_array)

ones_array = np.ones((4, 2))
print("2. All ones array (4x2):\n", ones_array)

full_array = np.full((3, 3), 7)
print("3. Full array (3x3) with 7:\n", full_array)

ref_array = np.array([[1, 2], [3, 4]])
zeros_like_array = np.zeros_like(ref_array)
print("4. Zeros like ref_array:\n", zeros_like_array)

ones_like_array = np.ones_like(ref_array)
print("5. Ones like ref_array:\n", ones_like_array)

new_list = [1, 2, 3, 4]
converted_array = np.array(new_list)
print("6. Converted list to array:\n", converted_array)
```

```python
import numpy as np

print("\nProblem 3: Array Operations")

x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

print("x:", x)
print("y:", y)
print("v:", v)
print("w:", w)

print("1. x + y:", x + y)
print("2. x - y:", x - y)
print("3. x * 3:", x * 3)
print("4. Square of x:", x ** 2)

print("5. Dot products:")
print("v . w =", np.dot(v, w))
print("x . v =\n", np.dot(x, v))
print("x . y =\n", np.dot(x, y))

print("6. Concatenate x and y along rows:\n", np.concatenate((x, y), axis=0))
print("Concatenate v and w along columns:\n", np.vstack((v, w)))

print("7. Concatenate x and v:")
try:
    result = np.concatenate((x, v), axis=0)
    print(result)
except ValueError as e:
    print("Error:", e)
```

```python
import numpy as np

print("Problem 4: Matrix Operations")

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

A_inv = np.linalg.inv(A)
identity_check = np.dot(A, A_inv)
print("1. A * A_inv:", identity_check)

AB = np.dot(A, B)
BA = np.dot(B, A)
print("2. AB:", AB)
print("BA:", BA)
print("AB == BA?", np.allclose(AB, BA))

AB_T = AB.T
BT_AT = np.dot(B.T, A.T)
print("3. (AB)^T:", AB_T)
print("B^T * A^T:", BT_AT)
print("(AB)^T == B^T * A^T?", np.allclose(AB_T, BT_AT))
```

```python
import numpy as np

print("Solving Linear Equations")

A_sys = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3, 1, -1]])
B_sys = np.array([-1, -3, 9])

A_inv_sys = np.linalg.inv(A_sys)
X = np.dot(A_inv_sys, B_sys)

print("Solution [x, y, z]:", X)
```
